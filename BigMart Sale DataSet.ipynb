```python
def check_outliers(x):
    q1 = np.percentile(x,25)
    q3 = np.percentile(x,75)
    lower = q1 - ((q3-q1)*1.5)
    upper = q1 + ((q3-q1)*1.5)
    return list(filter(lambda a:lower<=a<=upper,x))
def normalize(x):
    mu = np.mean(x)
    sd = np.std(x)
    b=[]
    t =[]
    for i in x:
        t.append(i)
    t.sort()
    check_normal(t)
    for i in x:
        b.append((i-mu)/sd)
    b.sort()
    #check_normal(b)
    a = check_outliers(b)
    #check_normal(a)
    plt.plot(a,norm.pdf(a))
    plt.show()
#scatter plot
def check_normal(x):
    p=[(x.index(i)-0.5)/len(x) for i in x]
    z=zscore(p)
    plt.scatter(x,p)
    plt.plot()
    plt.show()

normalize(df.Item_MRP)
```

### Hypothesis Testing

#### H0: For the item cost,  mean>=150

#### H1: For the item cost, mean<150

#### Take A Random Sample And Perform Z Test

```python
import random
l=random.sample(set(df.Item_MRP),150)
```

```python
xbar=sum(l)/len(l)
print(xbar)
```

```python
munot=150
```

```python
n=150
```

```python
import math
l2=list(map(lambda x:(x-xbar)**2,l))
sigma=sum(l2)/(n-1)
sigma=math.sqrt(sigma)
print(sigma)
```

```python
z=(xbar-munot)/(sigma/(math.sqrt(n)))
print(z)
```

```python
import scipy
p_values = scipy.stats.norm.sf(abs(z))
```

```python
print(p_values)
```

### Keeping a 5% statistical significance, alpha=0.05
### In this Trial ( Note the p-value for this random sample,changes for every execution of code since sample is random) 
### i)if(p-value>alpha)
### {
### Thus p-value does not give enough evidence against NULL hypothesis.
### Now , Outcome implies , H0 is plausible.
### }
### ii) else
### {
### NULL hypothesis is rejected.
### }
